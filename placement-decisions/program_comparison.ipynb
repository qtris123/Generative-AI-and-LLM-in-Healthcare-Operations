## Divide 'all_codes' into 2 categories ( value combinations with same setting number vs different setting number)
'all_codes' is a dataframe format of 'condition_all_all'

```python
same_setting = pd.DataFrame( columns=['code', 'values'])
diff_setting = pd.DataFrame( columns=['code', 'values'])

for index in range(len(all_codes)):
    anchor = all_codes.iloc[index]['values'][0][ len(all_codes.iloc[index]['values'][0]) - 1:]
    check = 'same'
    for program in all_codes.iloc[index]['values']:
        if program[len(program) - 1: ] != anchor:
            check = 'diff'
            break
    if check == 'same':
        same_setting.loc[len(same_setting)] = [all_codes.loc[index]['code'], all_codes.loc[index]['values']]
    elif check == 'diff':
        diff_setting.loc[len(diff_setting)] = [all_codes.loc[index]['code'], all_codes.loc[index]['values']]
```

```python
codes_diff_programs = pd.DataFrame( columns=['code', 'values'] )

for key in condition_all_all.keys():
    anchor = condition_all_all[key][0][:len(condition_all_all[key][0]) - 2]
    for program in condition_all_all[key]:
        if program[:len(program) - 2] != anchor:
            codes_diff_programs.loc[len(codes_diff_programs)] = [key, condition_all_all[key]]
            break
```
